```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os.path
```

```python
def fig_input(dataroot):
    
    # loading
    image = cv2.imread(dataroot)
    assert image is not None, "file could not be read, check the path"     
 
    # convertion to RGB
    image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # image's name and format
    name = (str(dataroot)).split('/')[-1]
    fig_name, file_format = name.split('.')
    print(name)
    
    #print
    plt.imshow(image_RGB)
    plt.show()
    
    return image_RGB, fig_name, file_format
```

```python
# images input

dataroot1 = 'C:/Users/MyWork/PycharmProjects/HW59_Linear_Filter/pani.jpg'
dataroot2 = 'C:/Users/MyWork/PycharmProjects/HW56_Color_Models/cat.jpg'

img1, fig_name1, file_format1 = fig_input(dataroot1)
img2, fig_name2, file_format2 = fig_input(dataroot2)
```

```python
# Sobel filter

def sobel_filter(img):

    # grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    grey = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    kernel_x = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
    kernel_y = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])

    edges_x = cv2.filter2D(grey, -1, kernel_x)
    edges_y = cv2.filter2D(grey, -1, kernel_y)
    edges = edges_x + edges_y
    
    plt.imshow(edges, cmap='gray')
    plt.show()
```

```python
img_sobel1 = sobel_filter(img1)
img_sobel2 = sobel_filter(img2)
```
